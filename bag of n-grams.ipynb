```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = "Virat Kohli is an Indian international cricketer who plays Test and ODI cricket for the Indian national team. A former captain in all formats of the game, Kohli retired from the T20I format following India's win at the 2024 T20 World Cup."
word = CountVectorizer(ngram_range=(1, 1))
word.fit_transform([text])
word.vocabulary_
```

```python
word1 = CountVectorizer(ngram_range=(1, 2))
word1.fit_transform([text])
word1.vocabulary_
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')
def preprocess(text):
    doc = nlp(text)
    words = [i.lemma_ for i in doc if not i.is_stop if not i.is_punct]
    return " ".join(words)
```

```python
preprocess("thor ate pizza")
```

```python
preprocess(text)
```

```python
vec = CountVectorizer(ngram_range=(1, 2))
vec.fit([preprocess(text)])
```

```python
dir(vec)
```

```python
vec.vocabulary_
```

```python
vec.fit([preprocess(text), preprocess("this is fun for me, the trip to kodaikanal is just awesome")])
```

```python
vec.transform([preprocess(text), preprocess("this is fun for me, the trip to kodaikanal is just awesome")]).toarray()
```

```python
df1 = df[df['category'].apply(lambda x: x in ['EDUCATION','ENVIRONMENT','CRIME','COMEDY','POLITICS'])]
```

```python
df1.shape
```

```python
df1_edu = df1[df1['category'] == 'EDUCATION'].sample(200, random_state=90)
df1_env = df1[df1['category'] == 'ENVIRONMENT'].sample(200, random_state=90)
df1_cri = df1[df1['category'] == 'CRIME'].sample(200, random_state=90)
df1_com = df1[df1['category'] == 'COMEDY'].sample(200, random_state=90)
df1_pol = df1[df1['category'] == 'POLITICS'].sample(200, random_state=90)
```

```python
df2 = pd.concat([df1_edu, df1_env, df1_cri, df1_com, df1_pol], axis = 'rows', ignore_index=True)
```

```python
df2.sample(2)
```

```python
df2.shape
```

```python
x = df2['short_description']
y = df2['category']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=90, stratify=y)
```

```python
x_train.shape, x_test.shape
```

```python
x_train = x_train.apply(lambda x: preprocess(x))
```

```python
x_test = x_test.apply(lambda x: preprocess(x))
```

```python
x_train
```

```python
y_train = y_train.replace({'POLITICS':0,'EDUCATION':1, 'ENVIRONMENT':2, 'CRIME':3, 'COMEDY':4})
```

```python
y_test = y_test.replace({'POLITICS':0,'EDUCATION':1, 'ENVIRONMENT':2, 'CRIME':3, 'COMEDY':4})
```

```python
y_train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

pipe = Pipeline([
    ('vector', CountVectorizer(ngram_range=(1, 2))),
    ('naive', MultinomialNB())
])
pipe.fit(x_train, y_train)
```

```python
pipe.score(x_test, y_test)
```

```python
pipe.predict(['the man killed him'])
```
