```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == "spam" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df['Message'], df['spam'], test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

text_to_num = CountVectorizer()
x = text_to_num.fit_transform(x_train)
```

```python
x.toarray()
```

```python
dir(text_to_num)
```

```python
text_to_num.vocabulary_           # this vocabulary show the position of the words
```

```python
x = x.toarray()
```

```python
x[0][x[0] != 0]
```

```python
np.where(x[0]!=0)
```

```python
from sklearn.naive_bayes import MultinomialNB

clas = MultinomialNB()
clas.fit(x, y_train)
```

```python
x_test = text_to_num.transform(x_test)
```

```python
x_test = x_test.toarray()
```

```python
clas.score(x_test, y_test)
```

```python
df['analysis'] = df['sentiment'].apply(lambda x: 0 if x == 'positive' else 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df['review'], df['analysis'], test_size=0.2, random_state=90)
```

```python
x_train.shape, y_train.shape, x_test.shape
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('countvector', CountVectorizer()),
    ('bayes algo', MultinomialNB())
])
```

```python
pipe.fit(x_train, y_train)
```

```python
pipe.score(x_test, y_test)
```
